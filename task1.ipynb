```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("population.csv")

# Check first rows
print(df.head())

# Select top 10 countries by population in 2020
top10 = df[['Country Name', '2020']].dropna()

# Sort values
top10 = top10.sort_values(by='2020', ascending=False).head(10)

# Create bar chart
plt.figure(figsize=(12,6))
plt.bar(top10['Country Name'], top10['2020'])

# Rotate labels
plt.xticks(rotation=45)

# Labels and title
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Countries by Population (2020)")

# Show chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
df = pd.read_csv("train.csv")

# Display first 5 rows
print(df.head())

# Dataset information
print(df.info())

# Check missing values
print(df.isnull().sum())

# Data Cleaning

# Fill missing Age values with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill missing Embarked values with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin column because it has many null values
df.drop(columns=['Cabin'], inplace=True)

# Check missing values again
print(df.isnull().sum())

```

```python
# Survival Count
plt.figure(figsize=(6,4))
df['Survived'].value_counts().plot(kind='bar')

plt.xlabel("Survived")
plt.ylabel("Count")
plt.title("Survival Count")
plt.show()
```

```python
# Gender vs Survival
plt.figure(figsize=(6,4))
pd.crosstab(df['Sex'], df['Survived']).plot(kind='bar')

plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender vs Survival")
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=20)

plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
# Passenger Class Distribution
plt.figure(figsize=(6,4))
df['Pclass'].value_counts().plot(kind='bar')

plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Passenger Class Distribution")
plt.show()
```
